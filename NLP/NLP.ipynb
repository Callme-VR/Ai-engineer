```python
def remove_number(text):
    new = ''
    for i in text:
        if not i.isdigit():
            new += i
    return new

data_frame['text'] = data_frame['text'].apply(remove_number)
```

```python
def remove_emojies(text):
    new =''
    for char in text:
        if char.isascii():
            new += char
    return new

data_frame['clean_text'] = data_frame['text'].apply(remove_emojies)
```

```python
from nltk.tokenize import word_tokenize

def removestopwords(text):
    words = word_tokenize(text)
    cleaned_text = []
    
    for i in words:
        if i.lower() not in stop_words:   # üëà key fix
            cleaned_text.append(i)
            
    return ' '.join(cleaned_text)
```

```python
data_frame['text'] = data_frame['text'].apply(removestopwords)

data_frame.loc[1]['text']
```
