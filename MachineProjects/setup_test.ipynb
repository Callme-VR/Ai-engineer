```python
#  after the data cleaning and  anltics and preprocessing  of data with better visuality
# we need to make the and use the NLP for the tokenization of data in word using  ------> two Techniques (Limitization,staming)

# NLP


import nltk
from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer 
import re
# re refer to regular expression  useful for puntuation 
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocessor_text(text):
    # convert to string and lowercase
    text = str(text).lower()
    
    # remove punctuation and numbers
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    
    # tokenize
    words = text.split()
    
    # remove stopwords
    words = [word for word in words if word not in stop_words]
    
    # lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]
    
    # join words back into sentence
    return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocessor_text)
df['tags'][0]
```
